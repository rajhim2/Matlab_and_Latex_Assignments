```python
def unipolar(inp):
	inp1=list(inp)
	li=[]
	li.append(0)
	for i in range(len(inp1)):
		if inp1[i]==1:
			li.append(2)
		else:
			li.append(0)
	return li
```

```python
def polar_nrz_l(inp):
    inp1=list(inp)
    inp1.insert(0,0)
    inp1=[-1 if i==0 else 1 for i in inp1]
    return inp1
```

```python
def polar_nrz_i(inp):
	inp2=list(inp)
	li=[]
	lock=False
	li.append(-1)
	for i in range(len(inp2)):
		if not lock:
			if inp2[i]==0:
				li.append(-1)
				lock=True
				continue
			else:
				li.append(1)
				lock=True
				continue
		if lock and inp2[i]==1:
			if li[i]==-1:
				li.append(1)
				continue
			else:
				li.append(-1)
				continue
		if lock and inp2[i]==0:
			li.append(li[i])
			continue
	return li
```

```python
def polar_rz(inp):
    inp1=list(inp)
    inp1=[-1 if i==0 else 1 for i in inp1]
    li=[]
    li.append(0)
    for i in range(len(inp1)):
        li.append(inp1[i])
        li.append(0)
    return li
```

```python
def Biphase_manchester(inp):
    inp1=list(inp)
    li=[]
    li.append(0)
    for i in range(len(inp1)):
        if inp1[i]==0:
            li.append(1)
            li.append(-1)
        elif inp1[i]==1 :
            li.append(-1)
            li.append(1)
    return li        
    
```

```python
def Differential_manchester(inp):
    inp1=list(inp)
    li,pre=[],''
    li.append(0)
    if inp1[0]==1:
        li.append(-1)
        li.append(1)
        pre='S'
    else:
        li.append(1)
        li.append(-1)
        pre='Z'
    for i in range(1,len(inp1)):
            if inp1[i]==0:
                if pre=='S':
                    li.append(-1);li.append(1)
                else:
                    li.append(1);li.append(-1)
            else:
                if pre=='Z':
                    pre='S'
                    li.append(-1);li.append(1)
                else:
                    pre='Z'
                    li.append(1);li.append(-1)
                         
    return li 
```

```python
def AMI(inp):
    inp1=list(inp)
    inp1.insert(0,0)
    lock=False
    for i in range(len(inp1)):
        if inp1[i]==1 and not lock:
            lock=True
            continue
        elif lock and inp1[i]==1:
            inp1[i]=-1
            lock=False
    return inp1  
```

```python
def Pseudoternary(inp):
    inp1=list(inp)
    inp1.insert(0,0)
    lock=False
    for i in range(1,len(inp1)):
        if inp1[i]==0 and not lock:
            inp1[i]=-1
            lock=True
            continue
        elif lock and inp1[i]==0:
            inp1[i]=1
            lock=False
        elif inp1[i]==1:
            inp1[i]=0
    return inp1
```

```python
def plot(li):
    plt.figure(figsize=(15,21), dpi=200)
    #plt.subplots(figsize=(13,13), dpi=100)
    plt.subplot(9,1,1)
    plt.ylabel("Unipolar-NRZ")
    plt.plot(unipolar(li),color='red',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,2)
    plt.ylabel("Polar-NRZ-L")
    plt.plot(polar_nrz_l(li),color='blue',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,3)
    plt.ylabel("Polar-NRZ-I")
    plt.plot(polar_nrz_i(li),color='green',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,4)
    plt.ylabel("Polar-RZ")
    plt.plot(polar_rz(li),color='red',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,5)
    plt.ylabel("Biphase-Manchester")
    plt.plot(Biphase_manchester(li),color='violet',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,6)
    plt.ylabel("Differential_Manchester")
    plt.plot(Differential_manchester(li),color='red',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,7)
    plt.ylabel("A-M-I")
    plt.plot(AMI(li),color='blue',drawstyle='steps-pre',marker='>')
    plt.subplot(9,1,8)
    plt.ylabel("Pseudoternary")
    plt.plot(Pseudoternary(li),color='blue',drawstyle='steps-pre',marker='>')
    plt.show()
    
```
